````python
def initialize_inventory(products):
    """Creates a dictionary to store the inventory"""
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory

#Modify the `get_customer_orders` function to include error handling.
#- If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), 
#display an error message and ask them to re-enter the number of orders.
#- If the user enters an invalid product name (e.g., a product name that is not in the inventory), 
#or that doesn't have stock available, display an error message and ask them to re-enter the product name. 
#*Hint: you will need to pass inventory as a parameter*
#- Use a try-except block to handle the error and continue prompting the user until a valid product name 
#is entered.

def get_customer_orders(inventory):
    """Groups the items ordered by the costumer"""
    customer_orders = set()
    valid_input = False   
    while valid_input == False:
        try:
            number_orders = int(input("Enter the number of customer orders: "))
            if number_orders >= 0:
                valid_input = True
            else: 
                print("The number of customer orders cannot be negative. Please enter a valid number of customer orders.")
        except ValueError:
            print("Invalid input. Please enter a number.")
    for i in range(number_orders):
        valid_product = False 
        while not valid_product:
            product = input("Enter the name of a product that a customer wants to order: ")
            try:              
                if product in inventory and inventory[product] > 0:
                    customer_orders.add(product)
                    valid_product = True
                else:
                    if product not in inventory:
                        raise ValueError(f"Error: {product} is not available in the inventory.")
                    else:
                        raise ValueError(f"Error: {product} is out of stock.")                
            except ValueError as e:
                print(str(e))
    return customer_orders

#Modify the calculate_total_price function to include error handling.
#If the user enters an invalid price (e.g., a negative value or a non-numeric value), display an error message 
#and ask them to re-enter the price for that product.
#Use a try-except block to handle the error and continue prompting the user until a valid price is entered.

def calculate_total_price(customer_orders):
    """Calculates the total price of the customer's order.
    Assumes only a maximum of 1 unit can be ordered."""
    total_price = 0
    for product in customer_orders:
        valid_input = False
        while valid_input == False:
            try:
                unit_price = int(input(f"Enter the unitary price of the {product}: "))
                if unit_price >= 0:
                    total_price = total_price + unit_price
                    valid_input = True
                else:
                    raise ValueError("Invalid price! Please enter a non-negative value.")
            except ValueError as error:
                print(f"Error: {error}")
    return total_price

def update_inventory(customer_orders, inventory):
    """Modifies the inventory according to the customer's orders"""
    inventory = {product: inventory[product] - 1 if product in customer_orders else inventory[product] for product in inventory}
    inventory = {product: amount for product, amount in inventory.items() if amount > 0}
    return inventory       
                                 
def calculate_order_statistics(customer_orders, products):
    """Calculates the total amount and percentage of unique products ordered.""" 
    TOP = len(customer_orders) #calculates the total amount of ordered products
    POP = len(customer_orders)*100/len(products) #calculates the percetage of unique products ordered
    return TOP,POP
#
def print_order_statistics(order_statistics):
    """Displays the order statistics"""
    print("Order Statistics: ")
    print(f"Unique products ordered: {order_statistics[0]}")
    print(f"Percentage of unique products ordered: {order_statistics[1]}")
#    
def print_updated_inventory(inventory):
    """Displays the updated inventory""" 
    print("Updated inventory:")
    {print(f"{product}: {amount}") for product, amount in inventory.items()}
                                 
                                 
````

````python
products = ["t-shirt", "mug", "book", "hat", "keychain"]
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)
order_statistics = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_statistics)
inventory = update_inventory(customer_orders, inventory)
print_updated_inventory(inventory)
total_price = calculate_total_price(customer_orders)
print(f"The total price of the customer order is : {total_price}")
````
